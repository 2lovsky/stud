1.Скачайте файл с предложениями (sentences.txt).

```python
text = open("sentences.txt", "r").read()
text
```

2.Считайте строки, приведите каждую к нижнему регистру с помощью строковой функции lower().

```python
txt = text.lower()
sentences = txt.split('\n')
sentences = sentences[0:-1]
sentences
```

3.Произведите токенизацию, то есть разбиение текстов на слова. 

```python
import re
words = re.split('[^a-z]', txt)
words = list(filter(lambda a: a != "", words))
words
```

4.Составьте список всех слов, встречающихся в предложениях. 

```python
wordlist = sorted(list(set(words))) # set работает как Union
len(wordlist)
```

```python
def list_words(text):
    splitted = re.split('[^a-z]', text)
    removed = list(filter(lambda a: a != "", splitted))
    out = sorted(list(set(removed)))
    return out
```

```python
list_words(txt)
```

Cопоставьте каждому слову индекс от нуля до (d - 1), где d — число различных слов в предложениях.

```python
my_dict = dict(zip(wordlist,range(0,len(wordlist))))
my_dict
```

5.Создайте матрицу размера n * d, где n — число предложений.

```python
import numpy as np
arr = np.zeros((len(sentences), len(wordlist)))
arr.shape
```

```python
my_dict[list_words(sentences[0])[0]]
```

```python
for i in range(len(list_words(sentences[0]))):
    num = my_dict[list_words(sentences[0])[i]]
    print(arr[(0, num)])
```

Заполните ее: элемент с индексом (i, j) в этой матрице 
должен быть равен количеству вхождений j-го слова в i-е предложение. 

```python
for i in range(len(sentences)):
    for j in range(len(list_words(sentences[i]))):
        num = my_dict[list_words(sentences[i])[j]]
        arr[i, num]+=1
```
